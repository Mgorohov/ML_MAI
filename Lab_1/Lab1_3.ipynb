3. Напишите функцию, которая принимает матрицу и производит нормировку её элементов (т.е. $A'_{ij} = A_{ij} / \sum_{i} \sum_{j} A_{ij}$)

```python
def normalizeMatrix(matrix):
    arr = np.array(matrix)
    if np.sum(arr) == 0:
        return arr
    newMatrix = matrix/np.sum(arr)
    return newMatrix

matrix = [[1, 2, 3],
          [2, 3, 4],
          [3, 4, 5]]
n_matrix = normalizeMatrix(matrix)
print(n_matrix)
```

4. Напишите функцию, которая принимает 3 трехмерных вектора $x, y, z$ и вычисляет объем пирамиды, образованной этими векторами (используйте смешанное произведение векторов)

```python
def volume(a, b, h):
    a = np.array(a)
    b = np.array(b)
    h = np.array(h)
    result = abs(np.dot(a, np.cross(b, h)))/6.0
    return result

a = [1, 2, 3]
b = [1, 3, 1]
h = [1, 4, 7]
v = volume(a, b, h)
print(v)
```

5. Напишите функцию, которая принимает число N и строит матрицу NxN, с нулями внутри и единицами на границе.

```python
def createMatrix(n):
    matrix = np.zeros((n, n))
    matrix[0, :] = 1
    matrix[-1, :] = 1
    matrix[:, 0] = 1
    matrix[:, -1] = 1
    return matrix

matr = createMatrix(5)
print(matr)
```
